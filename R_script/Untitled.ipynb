```python
df_m = pd.read_csv('./gene_module.csv', header = None)
```

```python
from collections import Counter

Counter(df_m[1]).most_common()
```

```python
dfg = pd.read_csv('../processed_data/adj_expression/Adipose_Subcutaneous-adj_expression.txt.gz')
```

```python
sns.clustermap(dfg[df_m[(df_m[1] == 'bisque4') | (df_m[1] == 'ivory')| (df_m[1] == 'brown4')][0]], cmap = 'RdBu_r', vmax = 6.0, vmin=-6.0)
plt.show()
```

```python
from sklearn.decomposition import PCA
clf = PCA(n_components = 2)
Xre = clf.fit_transform(dfg[df_m[(df_m[1] == 'bisque4') | (df_m[1] == 'ivory')| (df_m[1] == 'brown4')][0]].values)
```

```python
plt.plot(Xre[:, 0], Xre[:, 1], '.')
plt.show()
```
